**Apply functions to a dataframe df.apply() and df.applymap()**

* df.apply()

```python
df1_applied = df1.copy()

def custom_function(row):
    Country = row['Country']
    Action = row['Action']

    # Extract columns for summing (2018 - 2019 to 2023 - 2024)
    columns_to_sum = row.index[1:-2]

    # Convert columns to numeric and replace NaN values with 0
    numeric_values = pd.to_numeric(row[columns_to_sum].str.replace(',', ''), errors='coerce').fillna(0)

    total_sum = numeric_values.sum()

    # Format the total_sum with commas
    formatted_total_sum = f"{total_sum:,.0f}"

    result = f"{Country} total numeric values for {Action} from 2018 - 2024 is {formatted_total_sum} tons"

    return result

df1_applied['Total'] = df1_applied.apply(custom_function, axis=1)
df1_applied
```

**clean the text and nested json in**
> https://raw.githubusercontent.com/nnqomariyah/Fundamentals_of_Data_Science/main/week_3/data/movies_metadata.csv

```python
import json
```

```python
url = 'https://raw.githubusercontent.com/nnqomariyah/Fundamentals_of_Data_Science/main/week_3/data/movies_metadata.csv'
df3 = pd.read_csv(url)
```

*   Before cleaning the text data

```python
df3.info()
df3
```

*   Cleaning

```python
def clean_text(text):
    if pd.notna(text): # Skip NaN values
        cleaned_text = re.sub(r'\s+', ' ', text) # Remove extra whitespace and replace with a single space
        return cleaned_text
    else:
        return text

columns_to_clean = ['original_title', 'overview', 'tagline', 'title', 'tagline']

for column in columns_to_clean:
    df3[column] = df3[column].apply(clean_text)

# 'belongs_to_collection' column
def clean_belongs_to_collection(collection):
    if pd.notna(collection):
        match = re.search(r"'name':\s*'([^']+)'", collection) # Extract the 'name' field
        if match:
            return match.group(1)
    return None

df3['belongs_to_collection'] = df3['belongs_to_collection'].apply(clean_belongs_to_collection)

# 'production_companies', 'production_countries' columns
def clean_column(column):
    if pd.notna(column):
        try:
            data = json.loads(column.replace("'", "\""))
            if isinstance(data, list):
                item_list = [item['name'] for item in data]
                return ', '.join(item_list)
        except (json.JSONDecodeError, KeyError):
            return None
    return None

df3['production_companies'] = df3['production_companies'].apply(clean_column)
df3['production_countries'] = df3['production_countries'].apply(clean_column)

# 'runtime' column
def format_runtime(runtime):
    if pd.notna(runtime):
        return f'{int(runtime)} mins'
    return None

df3['runtime'] = df3['runtime'].apply(format_runtime)

# 'genres' column
df3['genres'] = df3['genres'].apply(lambda x: eval(x) if pd.notna(x) else []) # Convert to a list of dictionaries
df3['genres'] = df3['genres'].apply(lambda x: ', '.join(item['name'] for item in x))

# 'spoken_languages' column
df3['spoken_languages'] = df3['spoken_languages'].apply(lambda x: eval(x) if pd.notna(x) else [])
df3['spoken_languages_names'] = df3['spoken_languages'].apply(lambda x: ', '.join(item['name'] for item in x))
df3 = df3.drop(columns=['spoken_languages']) # Drop original column

# 'budget' column
def clean_budget(budget):
    if pd.notna(budget):
        numeric_budget = re.sub(r'[^\d]+', '', budget) # Remove non-numeric characters
        return numeric_budget
    return None

df3['budget'] = df3['budget'].apply(clean_budget)

# Add commas for numbers
def format_with_commas(number):
    if pd.notna(number):
        return '{:,}'.format(int(number))
    return None

df3['budget'] = df3['budget'].apply(format_with_commas)
df3['revenue'] = df3['revenue'].apply(format_with_commas)

# 'vote_average' column
df3['vote_average'] = df3['vote_average'].apply(lambda x: float(x) if pd.notna(x) else None)

# 'vote_count' column
df3['vote_count'] = df3['vote_count'].apply(lambda x: f'{int(x):,}' if pd.notna(x) else None)

# Dropping NaN values
df3.dropna(inplace=True)

# Drop 'adult', 'video', 'original_language' columns
df3.drop(columns=['adult', 'video', 'original_language'], inplace=True)

# Rearranging columns
columns_order = [
    'id',
    'imdb_id',
    'title',
    'original_title',
    'spoken_languages_names',
    'belongs_to_collection',
    'release_date',
    'genres',
    'tagline',
    'overview',
    'runtime',
    'production_companies',
    'production_countries',
    'budget',
    'revenue',
    'vote_average',
    'vote_count',
    'popularity',
    'homepage',
    'status',
    'poster_path'
]
df3 = df3[columns_order]

# Renaming columns
columns_rename = {
    'id': 'ID',
    'imdb_id': 'IMDB ID',
    'title': 'Title',
    'original_title': 'Original Title',
    'spoken_languages_names': 'Spoken Languages',
    'belongs_to_collection': 'Collection',
    'release_date': 'Release Date',
    'genres': 'Genres',
    'tagline': 'Tagline',
    'overview': 'Overview',
    'runtime': 'Runtime',
    'production_companies': 'Production Companies',
    'production_countries': 'Production Countries',
    'budget': 'Budget',
    'revenue': 'Revenue',
    'vote_average': 'Vote Average',
    'vote_count': 'Vote Count',
    'popularity': 'Popularity',
    'homepage': 'Homepage',
    'status': 'Status',
    'poster_path': 'Poster Path'
}

# Rename the columns in the DataFrame
df3.rename(columns = columns_rename, inplace = True)
```
